```python
import numpy as np
import pandas as pd
```

```python
#open sales csv
yx = pd.read_csv('../TSPerf/retail_sales/OrangeJuice_Pt_3Weeks_Weekly/data/data/yx.csv')
yx
```

```python
store_list = sorted(list(set(yx.iloc[:, 0])))
brand_list = sorted(list(set(yx.iloc[:, 1])))
```

```python
count = 1
tsMap = dict() #map from (store, brand) to 
for i in store_list: 
    for j in brand_list: 
        tsMap[(i, j)] = 'E' + str(count)
        count += 1
tsMap
```

```python
yx.insert(loc=0, column='SalesID', value=[float('nan')] * len(yx))   
```

```python
yx
```

```python
for i, val in yx.iterrows(): 
    idtuple = (int(val['store']), int(val['brand']))
    sales_id = tsMap[idtuple]
    yx.iloc[i, 0] = sales_id
yx
```

```python
yx.to_csv('../TSPerf/retail_sales/OrangeJuice_Pt_3Weeks_Weekly/data/data/yxmod.csv', index=False)
```
